```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
df1 = pd.read_csv("langs_by_county.csv") 
df1.head(3)
```

```python
df2 = pd.read_csv("state_code.csv") 
df2.head(3)
```

```python
combined_df = pd.merge(df1, df2, how='outer', on='state code')
combined_df.head()
```

```python
# French Percentage by County (TOP 30 COUNTIES THAT SPEAK FRENCH MOST FREFREQUENTLY)
dff_french = (combined_df.sort_values(by='pct_french', ascending=False).head(100))
dff_french
```

```python
fig, ax = plt.subplots(figsize=(100, 20))
ax.barh(dff_french['state name'], dff_french['pct_french'])
```

```python
# Set a Title and labels
plt.title("Percentage of Households Speaking French per State")
plt.xlabel("Percentage of Households Speaking French")
plt.ylabel("State")
```

```python
# Save our graph and show the graph
plt.tight_layout()
plt.savefig("bar_pct_french_state.png")
plt.show()
```
