```python
import pandas as pd
import numpy as np
import requests
import json
import csv
from apikey import api_key
```

```python
#https://api.census.gov/data/2013/language?get=EST,LANLABEL,LAN39,NAME&for=county:020&in=state:02&LAN=625

numbers = np.arange(600,999)

#numbers = [645, 663, 777]

langlist = []

for value in numbers:
    try: 
        line = requests.get(f"https://api.census.gov/data/2013/language?get=EST,LANLABEL,LAN39&for=county:031&in=state:17&LAN={value}&key={api_key}").json()
        print(line)
        langlist.append(line)
    except:
        print(f"No response from server for {value}")

langlist = [item[1] for item in langlist]

df = pd.DataFrame(langlist, columns = ["Speakers", "Language", "Lang_39", "Lang_Code", "State", "County"])

df.head()
```

```python
df.to_csv('cook_languages.csv', index=True)
```
