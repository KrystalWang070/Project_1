```python
import pandas as pd
import requests
import json
import csv
from config import api_key
```

```python
#sets up lists for each of the columns in the API request
est = [] #population estimate
lanlabel = [] #name of language
name = [] #name of the place
lan = [] #language code
state = [] #state code for FIPS
county = [] #county code for FIPS

#These are the language codes used by the Census in their API requests.

Spanish = 625
Arabic = 777
Chinese = 708
Mandarin = 712
French = 620
Polish = 645

#This is the census API URL:

url = "https://api.census.gov/data/2013/language?get=EST,LANLABEL,NAME&for=county:*&LAN="

#This is a sample built URL for Chinese

query_url = f"{url}{Chinese}&key={api_key}"

print(query_url)
```

```python
response = requests.get(query_url).json()

response_df = pd.DataFrame(response)

response_df.columns = response_df.iloc[0]
response_df = response_df.drop(response_df.index[0])

response_df.head()
```

```python
response_df["FIPS Code"] = response_df["state"].map(str) + response_df["county"]
#https://stackoverflow.com/questions/19377969/combine-two-columns-of-text-in-dataframe-in-pandas-python

#response_df = response_df.set_index("FIPS Code")

response_df.head()
```

```python
gps_coords = "County GPS Coordinates.csv"

latlong = pd.read_csv(gps_coords, converters={"FIPS": lambda x: str(x)})
#https://stackoverflow.com/questions/13250046/how-to-keep-leading-zeros-in-a-column-when-reading-csv-with-pandas

latlong_df = pd.DataFrame(latlong, dtype=str)

latlong_df = latlong_df.rename(columns={"FIPS": "FIPS Code"})

latlong_df.head()
```

```python
countymap_df = pd.merge(response_df, latlong_df, on = "FIPS Code")

countymap_df.head()
```
