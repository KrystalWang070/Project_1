```python

# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#read percentage file
bar_df = pd.read_csv("langs_by_county.csv") 
bar_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(bar_df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(40,3))
plt.bar(x_axis, bar_df["pct_french"], color='r', alpha=1, align="center")
plt.xticks(tick_locations, bar_df["county"], rotation="vertical")
```

```python
plt.xlim(-1, len(x_axis))
plt.ylim(0, max(bar_df["pct_french"])+5)
```

```python
# Set a Title and labels
plt.title("Percentage of Households Speaking French per State")
plt.xlabel("State")
plt.ylabel("Percentage of Households Speaking French")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("bar_pct_french.png")
plt.show()
```
